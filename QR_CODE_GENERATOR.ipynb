```python
import qrcode
#Taking UPI ID from the user
upi_id=input("Enter your UPI ID = ")

#upi://pay?pa=UPI_ID&pn=NAME&am=Amount&cu=CURRENCY&tn=MESSAGE

phonepay_url=f"upi://pay?pa={upi_id}&pn=Abhinav Sharma&mc=123"
paytm_url=f"upi://pay?pa={upi_id}&pn=Abhinav Sharma&mc=123"
googlepay_url=f"upi://pay?pa={upi_id}&pn=Abhinav Sharma&mc=123"

#create the qr code for different payment methods
phonepay_qr=qrcode.make(phonepay_url)
paytm_qr=qrcode.make(paytm_url)
googlepay_qr=qrcode.make(googlepay_url)

#save the qr into the image file
phonepay_qr.save("phonepay_qr.png")
paytm_qr.save("paytm_qr.png")
googlepay_qr.save("googlepay_qr.png")

#display qr code
print("For phone pay:    \n\n\n\n")
display(phonepay_qr)
print("For paytm:    \n\n\n\n")
display(paytm_qr)
print("For google pay:    \n\n\n\n")
display(googlepay_qr)
```
